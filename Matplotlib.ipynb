# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = {'Year':[1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020],
       'Exchange Rate':[65,69,71,64,62,59,72,71,75,78,81]}
df = pd.DataFrame(Data,columns = ['Year','Exchange Rate'])
```

```python
# df= pd.read_csv('filename.csv')
# df= pd.read_excel('filename.excel')
```

```python
df.plot(x='Year',y='Exchange Rate',kind='bar')
plt.show()
```

```python
df.plot(x='Year',y='Exchange Rate',kind='area')
plt.show()
```

```python
df.plot(x='Year',y='Exchange Rate',kind='line')
plt.show()
```

```python
df.plot(x='Year',y='Exchange Rate',kind='barh')
plt.show()
```

```python
df.plot(x='Year',y='Exchange Rate',kind='scatter')
plt.show()
```

```python
plt.scatter(df['Year'],df["Exchange Rate"]) #same as above it but different code
plt.show()
```

```python
labels = 'JavaScript','Java', 'Python','R'
sizes = [15,30,45,10]
explode_labels=(0,0.2,0.3,0)
fig1,ax1 = plt.subplots()

ax1.pie(sizes,explode=explode_labels,labels = labels,shadow = True,startangle = 90 )
ax1.axis('equal')
plt.show()
```

```python
labels = 'JavaScript','Java', 'Python','R'
sizes = [15,30,45,10]
explode_labels=(0,0.5,0,0)
fig1,ax1 = plt.subplots()

ax1.pie(sizes,explode=explode_labels,labels = labels,shadow = True,startangle = 90 )
ax1.axis('equal')
plt.show()
```
